```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
# ODE 1
def f1(t,y):
    return (t*np.exp(3*t))-(2*y)

def true_soln1(t):
    return (1/25)*(np.exp(-2*t)-np.exp(3*t)+5*(np.exp(3*t)*t))


t_arr1=np.linspace(0,1,20)
y_arr1=solve_ivp(f1,(0,1),[0],t_eval=t_arr1).y[0]
y_true1=true_soln1(t_arr1)
plt.plot(t_arr1,y_arr1,"--",color ='black',label="computed soln")
plt.plot(t_arr1,y_true1,color ='yellow',label="Exact Soln")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.title("ODE 1")
plt.show()

# ODE 2
def f2(t,y):
    return 1-((t-y)**2)

def true_soln2(t):
    return (1-(3*t)+(t**2))/(t-3)

t_arr2=np.linspace(2,2.9,20)
y_arr2=solve_ivp(f2,(2,3),[1],t_eval=t_arr2).y[0]
y_true2=true_soln2(t_arr2)
plt.plot(t_arr2,y_arr2,"--",color ='black',label="computed soln")
plt.plot(t_arr2,y_true2,color ='yellow',label="Exact Soln")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.title("ODE 2")
plt.show()

# ODE 3


def f3(t,y):
    return 1+(y/t)

def true_soln3(t):
    return 2*t+(t*np.log(t))

t_arr3=np.linspace(1,2,20)
y_arr3=solve_ivp(f3,(1,2),[2],t_eval=t_arr3).y[0]
y_true3=true_soln3(t_arr3)
plt.plot(t_arr3,y_arr3,"--",color ='black',label="computed soln")
plt.plot(t_arr3,y_true3,color ='yellow',label="Exact Soln")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.title("ODE 3")
plt.show()


# ODE 4


def f4(t,y):
    return np.cos(2*t)+np.sin(3*t)

def true_soln4(t):
    return (1/6)*(8-2*np.cos(3*t)+3*np.sin(2*t))

t_arr4=np.linspace(0,1,20)
y_arr4=solve_ivp(f4,(0,1),[1],t_eval=t_arr4).y[0]
y_true4=true_soln4(t_arr4)
plt.plot(t_arr4,y_arr4,"--",color ='black',label="computed soln")
plt.plot(t_arr4,y_true4,color ='yellow',label="Exact Soln")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.title("ODE 4")
plt.show()
```
