```python
import numpy as np
import matplotlib.pyplot as plt

def du(t,u):
    return (np.array([u[0]+2*u[1]-2*u[2]+np.exp(-t),u[1]+u[2]-2*np.exp(-t), u[0]+2*u[1]+np.exp(-t)]))

def ode_solve_RK4(f,t_arr,y0):
    y=np.zeros((len(t_arr),len(y0)))
    h=t_arr[1]-t_arr[0]
    y[0,:]=y0
    for i in range(len(t_arr)-1):
        k1 =h*f(t_arr[i],y[i,:])
        k2 =h*f(t_arr[i]+(h/2),y[i,:]+k1/2)
        k3 =h*f(t_arr[i]+(h/2),y[i,:]+k2/2)
        k4 =h*f(t_arr[i]+h,y[i,:]+k3)
        y[i+1,:]=y[i,:]+(1/6)*(k1+(2*k2)+(2*k3)+k4)
    return y

u0 = [3,-1,1]
t_arr = np.linspace(0,1,100)
soln = ode_solve_RK4(du,t_arr,u0)

for i in range(3):
    plt.plot(t_arr,soln[:,i],label = "u"+str(i+1))
    plt.xlabel("t")
    plt.ylabel("u(t)")
    plt.legend()

```
