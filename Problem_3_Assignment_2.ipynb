```python
import numpy as np
import matplotlib.pyplot as plt
# Define the function representing the second-order differential equation
def f(x, y, yp):
    return x * np.exp(x) - x - y + 2 * yp

# Implement the 4th-order Runge-Kutta method
def runge_kutta_4th_order(x0, y0, yp0, xn, h):
    # Number of steps
    n = int((xn - x0) / h)

    # Arrays to store x and y values
    x_values = [x0]
    y_values = [y0]

    # Initialize initial y' value
    yp = yp0

    # Runge-Kutta iteration
    for i in range(n):
        x = x_values[-1]
        y = y_values[-1]

        k1 = h * yp
        l1 = h * f(x, y, yp)
        
        k2 = h * (yp + 0.5 * l1)
        l2 = h * f(x + 0.5 * h, y + 0.5 * k1, yp + 0.5 * l1)
        
        k3 = h * (yp + 0.5 * l2)
        l3 = h * f(x + 0.5 * h, y + 0.5 * k2, yp + 0.5 * l2)
        
        k4 = h * (yp + l3)
        l4 = h * f(x + h, y + k3, yp + l3)

        y_new = y + (k1 + 2*k2 + 2*k3 + k4) / 6
        yp_new = yp + (l1 + 2*l2 + 2*l3 + l4) / 6

        x_values.append(x + h)
        y_values.append(y_new)
        yp = yp_new

    return x_values, y_values

# Define initial conditions and parameters
x0 = 0
y0 = 0
yp0 = 0
xn = 1
h = 0.01

# Solve the differential equation using the 4th-order Runge-Kutta method
x_values, y_values = runge_kutta_4th_order(x0, y0, yp0, xn, h)

# Print the results
#print("x values:", x_values)
#print("y values:", y_values)
plt.plot(x_values,y_values)
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("y''-2y'+2=x(e^x)-x using RK4 Method")
```
