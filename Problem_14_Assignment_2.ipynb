```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y, yp):
    return (t**3 * np.log(t) + 2 * t * yp - 2 * y) / t**2


def euler_method(t0, y0, yp0, tn, h):
    # Number of steps
    n = int((tn - t0) / h)

    # Arrays to store t, y, and yp values
    t_values = [t0]
    y_values = [y0]
    yp_values = [yp0]

    # Euler iteration
    for i in range(n):
        t = t_values[-1]
        y = y_values[-1]
        yp = yp_values[-1]

        y_next = y + h * yp
        yp_next = yp + h * f(t, y, yp)

        t_values.append(t + h)
        y_values.append(y_next)
        yp_values.append(yp_next)

    return t_values, y_values

t0 = 1
y0 = 1
yp0 = 0
tn = 2
h = 0.001

t_values, y_values = euler_method(t0, y0, yp0, tn, h)


def exact_solution(t):
    return 7 * t / 4 + (t**(3)/2) * np.log(t) - (3/4) * t**3

exact_values = exact_solution(np.array(t_values))

plt.plot(t_values, y_values, label='Numerical Solution (Euler\'s Method)', color='blue')
plt.plot(t_values, exact_values, label='Exact Solution', linestyle='--', color='red')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Comparison of Numerical and Exact Solutions')
plt.legend()
plt.grid(True)
plt.show()
```
