```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_bvp

# BVP1
def f1(x,y):
    return np.array([y[1],-np.exp(-2*y[0])])

def boundary_conditions1(ya,yb):
    return np.array([ya[0],yb[0]-np.log(2)])

x_values1=np.linspace(1,2,100)
y_guess1 = np.zeros((2, len(x_values1)))
sol1 = solve_bvp(f1, boundary_conditions1, x_values1, y_guess1)

def true_sol1(t):
    return np.log(t)


plt.plot(sol1.x,sol1.y[0],"--",color ='black',label="computed soln")
plt.plot(sol1.x,true_sol1(sol1.x),color ='yellow',label="Exact Soln")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.title("BVP 1")
plt.show()


# BVP 2

def f2(x, y):
    return np.array([y[1],(y[1]*np.cos(x))-(y[0]*np.log(y[0]))])

def boundary_conditions2(ya, yb):
    return np.array([ya[0]-1, yb[0] - np.exp(1)])  # y(a) = 1, y(b) = 1


x_values2 = np.linspace(0.001,np.pi/2, 100)
y_guess2 = np.ones((2, len(x_values2)))
sol2 = solve_bvp(f2, boundary_conditions2, x_values2, y_guess2)

def true_sol2(t):
    return np.exp(np.sin(t))

plt.plot(sol2.x,sol2.y[0],"--",color ='black',label="computed soln")
plt.plot(sol2.x,true_sol2(sol2.x),color ='yellow',label="Exact Soln")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.title("BVP 2")
plt.show()


# BVP 3


def f3(x, y):
    return np.array([y[1],-( (2*(y[1]**3)) + ((y[0]**2)*y[1]) )/ np.cos(x)])

def boundary_conditions3(ya, yb):
    return np.array([ya[0]-(2**(-0.25)), yb[0] - ((12**0.25)/2)])

def true_sol3(t):
    return np.sin(t)**0.5

x_values3 = np.linspace(np.pi/4,np.pi/3, 100)
y_guess3 = np.ones((2, len(x_values3)))
sol3 = solve_bvp(f3, boundary_conditions3, x_values3, y_guess3)

plt.plot(sol3.x,sol3.y[0],"--",color ='black',label="computed soln")
plt.plot(sol3.x,true_sol3(sol3.x),color ='yellow',label="Exact Soln")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.title("BVP 3")
plt.show()

# BVP 4


def f4(x, y):
    return np.array([y[1],0.5-((y[1]**2)/2)-((y[0]*np.sin(x))/2)])

def boundary_conditions4(ya, yb):
    return np.array([ya[0]-2, yb[0] - 2])

x_values4 = np.linspace(0,np.pi, 100)
y_guess4 = np.ones((2, len(x_values4)))
sol4 = solve_bvp(f4, boundary_conditions4, x_values4, y_guess4)

def true_sol4(t):
    return (2+np.sin(t))

plt.plot(sol4.x,sol4.y[0],"--",color ='black',label="computed soln")
plt.plot(sol4.x,true_sol4(sol4.x),color ='yellow',label="Exact Soln")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.title("BVP 4")
plt.show()
```
